### Задача 2
В первом ящике находится 8 мячей, из которых 5 - белые. Во втором ящике - 12 мячей, из которых 5 белых. Из первого ящика вытаскивают случайным образом два мяча, из второго - 4. Какова вероятность того, что 3 мяча белые?

```python
N1 = 8 # кол. мячей в 1-ом ящике
M1 = 5 # белых мячей в 1-ом ящике
N2 = 12 # кол. мячей во 2-м ящике
M2 = 5 # белых мячей во 2-ом ящике
K1 = 2 # кол. мячей, вытаскиваемых из 1-го ящика
K2 = 4 # кол. мячей, вытаскиваемых из 2-го ящика

total1 = cmb.combNum(N1, K1) # общее кол. способов вытащить мячи из 1-го ящика
total2 = cmb.combNum(N2, K2) # общее кол. способов вытащить мячи из 2-го ящика

# вероятности вытащить определённое кол. белых мячей
pList = [[], [0]] # [bn][i] bn: 0 - 1-ый ящик, 1 - 2-ой ящик; i - кол. вытащенных белых мячей (0 для 2-го ящика не рассматриваем)
for i in range(3):
    pList[0].append(cmb.combNum(M1, i) * cmb.combNum(N1 - M1, K1 - i) / total1)

for i in range(1, 4):
    pList[1].append(cmb.combNum(M2, i) * cmb.combNum(N2 - M2, K2 - i) / total2)

# складываем возможные ситуации
p = pList[0][0] * pList[1][3] + \
    pList[0][1] * pList[1][2] + \
    pList[0][2] * pList[1][1]

print(f"{round(p * 100, 2)}%")
```

### Задача 3
На соревновании по биатлону один из трех спортсменов стреляет и попадает в мишень. Вероятность попадания для первого спортсмена равна 0.9, для второго — 0.8, для третьего — 0.6. Найти вероятность того, что выстрел произведен: a). первым спортсменом б). вторым спортсменом в). третьим спортсменом.

```python
P1, P2, P3 = 0.9, 0.8, 0.6 # вероятности попадания 1-го, 2-го и 3-го спортсменов, соответственно

pFull = cmb.symFullProb(3, [P1, P2, P3]) # вероятность, что в мишень попадут
print("Вероятность, что стрелял 1-ый спортсмен: ", end="")
print(f"{round(cmb.bayes(p_ba=P1, p_a=1/3, p_b=pFull) * 100, 2)}%")
print("Вероятность, что стрелял 2-ый спортсмен: ", end="")
print(f"{round(cmb.bayes(p_ba=P2, p_a=1/3, p_b=pFull) * 100, 2)}%")
print("Вероятность, что стрелял 3-ий спортсмен: ", end="")
print(f"{round(cmb.bayes(p_ba=P3, p_a=1/3, p_b=pFull) * 100, 2)}%")
```

### Задача 4
В университет на факультеты A и B поступило равное количество студентов, а на факультет C студентов поступило столько же, сколько на A и B вместе. Вероятность того, что студент факультета A сдаст первую сессию, равна 0.8. Для студента факультета B эта вероятность равна 0.7, а для студента факультета C - 0.9. Студент сдал первую сессию. Какова вероятность, что он учится: a). на факультете A б). на факультете B в). на факультете C?

```python
P1, P2, P3 = 0.8, 0.7, 0.9 # вероятность сдачи сессии на факультетах A, B и C, соответственно
PA1, PA2, PA3 = 1/4, 1/4, 1/2 # вероятности, что студент учится на факультетах A, B и C, соответственно

pFull = cmb.symFullProb(4, [P1, P2, P3, P3]) # вероятность, что случайно взятый студент сдаст сессию
print("Вероятность, что студент учится на факультете A: ", end="")
print(f"{round(cmb.bayes(p_ba=P1, p_a=PA1, p_b=pFull) * 100, 2)}%")
print("Вероятность, что студент учится на факультете B: ", end="")
print(f"{round(cmb.bayes(p_ba=P2, p_a=PA2, p_b=pFull) * 100, 2)}%")
print("Вероятность, что студент учится на факультете C: ", end="")
print(f"{round(cmb.bayes(p_ba=P3, p_a=PA3, p_b=pFull) * 100, 2)}%")
```

### Задача 5
Устройство состоит из трех деталей. Для первой детали вероятность выйти из строя в первый месяц равна 0.1, для второй - 0.2, для третьей - 0.25. Какова вероятность того, что в первый месяц выйдут из строя: а). все детали б). только две детали в). хотя бы одна деталь г). от одной до двух деталей?

```python
P1, P2, P3 = 0.1, 0.2, 0.25 # вероятности выхода из строя 1-ой, 2-ой и 3-ей деталей, соответственно

p_0 = (1 - P1) * (1 - P2) * (1 - P3) #  вероятность, что ни одна деталь не вышла из строя
p_1 = P1 * (1 - P2) * (1 - P3) + (1 - P1) * P2 * (1 - P3) + (1 - P1) * (1 - P2) * P3 #  вероятность, что 1 деталь вышла из строя
p_2 = P1 * P2 * (1 - P3) + P1 * (1 - P2) * P3 + (1 - P1) * P2 * P3 #  вероятность, что 2 детали вышли из строя
p_3 = P1 * P2 * P3 # вероятность, что 3 детали вышли из строя

print(f"а: {round(p_3 * 100, 2)}%")
print(f"б: {round(p_2 * 100, 2)}%")
print(f"в: {round((1 - p_0) * 100, 2)}%")
print(f"г: {round((p_1 + p_2) * 100, 2)}%")
```
